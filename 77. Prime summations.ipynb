```
def sieveOfEratosthenes(n):
    """sieveOfEratosthenes(n): return the list of the primes < n."""
    # Code from: <dickinsm@gmail.com>, Nov 30 2006
    # http://groups.google.com/group/comp.lang.python/msg/f1f10ced88c68c2d
    if n <= 2:
        return []
    sieve = range(3, n, 2)
    top = len(sieve)
    for si in sieve:
        if si:
            bottom = (si*si - 3) // 2
            if bottom >= top:
                break
            sieve[bottom::si] = [0] * -((bottom - top) // si)
    return [2] + [el for el in sieve if el]
```

```
primes = sieveOfEratosthenes(10**6)
```

```
def sopf(n):
  total = 0
  for prime in primes:
    if n % prime == 0:
      total += prime
    
    if prime > n/2:
      if total != 0:
        return total
      else:
        return n
```

```
prime_partitions = [-1] * 100000
```

```
def prime_partition(n):
  """K(n) = sopf(n) + sigma(sopf(j)*K(n-j)) for 1<=j<=n-1"""
  if prime_partitions[n] != -1:
    return prime_partitions[n]
  else:
    sopf_n = sopf(n)
    total = 0
    total += sopf_n
    for j in xrange(1,n):
      sopf_j = sopf(j)
      summand = sopf * prime_partition(n-j)
      total += summand
  
  return 1.0/n * total
```

```
sum_of_prime_factors = {0:0, 1:0}
for key in xrange(2,100000):
  sum_of_prime_factors[key] = sopf(key)
```

```
kappas = {}
kappas[0] = 0
kappas[1] = 0
kappas[10] = 5
```

```
def kappa(n):
  """K(n) = 1/n(sopf(n) + sigma(sopf(j)*K(n-j))) for 1<=j<=n-1"""
  total = 0
  if n not in kappas:
    sopf_n_over_n = 1.0/n * sum_of_prime_factors[n]
    total += sopf_n_over_n
    for j in xrange(1,n):
      summand_over_n = (sum_of_prime_factors[j] * kappas[n-j]) * 1.0 / n
      total += summand_over_n
    
    kappas[n] = total
  
  return kappas[n]
```

```
for n in xrange(2,100000):
  kappas[n] = kappa(n)
  if kappas[n] > 5000:
    print "K(" + str(n) + ") = " + str(kappas[n])
    break
```
