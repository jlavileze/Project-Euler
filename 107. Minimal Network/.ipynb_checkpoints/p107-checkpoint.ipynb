```python
f = open("p107_network.txt", "r")
txt = f.readlines()
f.close()
```

```python
def convert_to_matrix(line):
    x1 = line.split(",")
    x2 = []
    for n in x1:
        try:
            x2.append(int(n))
        except:
            x2.append(np.nan)
    return x2
```

```python
matrix = list(map(convert_to_matrix, txt))
```
