```
def sieveOfEratosthenes(n):
    """sieveOfEratosthenes(n): return the list of the primes < n."""
    # Code from: <dickinsm@gmail.com>, Nov 30 2006
    # http://groups.google.com/group/comp.lang.python/msg/f1f10ced88c68c2d
    if n <= 2:
        return []
    sieve = range(3, n, 2)
    top = len(sieve)
    for si in sieve:
        if si:
            bottom = (si*si - 3) // 2
            if bottom >= top:
                break
            sieve[bottom::si] = [0] * -((bottom - top) // si)
    return [2] + [el for el in sieve if el]
```

```
def next_power(num):
    # num = 2^q - 1
    new_number = 2 * (num + 1) - 1
    return new_number

def differential(pre_tuple,post_tuple):
    power_diff = post_tuple[1] - pre_tuple[1]
    diff = pre_tuple[0] ** power_diff
    
    return diff
```

```
primes = sieveOfEratosthenes(7376508)
# primes = sieveOfEratosthenes(8)
primes_with_powers = []

for prime in primes:
    tup = (prime, 1)
    primes_with_powers.append(tup)
```

```
# This code gives the correct solution, but for some reason it is painfully slow
starting_point = 2

for counter, big_prime in enumerate(reversed(primes_with_powers)):
  if starting_point == 7376507:
    break
  for small_counter, prime_with_power in enumerate(primes_with_powers):
    print "candidate: " + str(prime_with_power[0]) + "Limit: " + str(starting_point)
    if prime_with_power[0] >= starting_point:
      if differential(prime_with_power, (prime_with_power[0], next_power(prime_with_power[1]))) < big_prime[0]:
        index = -1 * (counter + 1)
        primes_with_powers[index] = (primes_with_powers[index][0], 0)
        primes_with_powers[small_counter] = (primes_with_powers[small_counter][0], next_power(primes_with_powers[small_counter][1]))
        break
      elif small_counter + 1 < 500500:
        starting_point = primes_with_powers[small_counter + 1][0]
```
